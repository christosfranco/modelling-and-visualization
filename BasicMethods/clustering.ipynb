```python
from matplotlib import pyplot as plt 
import numpy as np
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
X = np.random.randint(100, size=(100, 3))
fig = plt.figure()
```

```python


ax = Axes3D(fig)

# Data for a three-dimensional line
ax.scatter(X[:,0],X[:,1],X[:,2])

plt.show()
```

```python
# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate the values
x_vals = X[:,0]
y_vals = X[:,1]
z_vals = X[:,2]

# Plot the values
ax.scatter(x_vals, y_vals, z_vals, c = 'b', marker='o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],X[:,2],label='True position')
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Generate the values
x_vals = X[:,0]
y_vals = X[:,1]
z_vals = X[:,2]

# Plot the values
ax.scatter(x_vals, y_vals, z_vals,  c = kmeans.labels_,cmap='rainbow')
ax.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,2], color='black')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,0],color='black')
```

```python
plt.scatter(X[:,1],X[:,2],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,2],color='black')
```
